```python
import collections
class Tnode:
    def __init__(self, value):
        self.value = value
        self.children = []
        self.parent = None
    
    def add_child(self, child):
        child.parent = self
        self.children.append(child)
    
    def get_level(self):
        level = 0
        current = self.parent
        while current:
            level += 1
            current = current.parent
        return level
    
    def print_tree(self):
        spaces = '\t' * self.get_level() 
        prefix = spaces + "|-- " if self.parent else ""
        print(prefix + str(self.value))
        if self.children:
            for child in self.children:
                child.print_tree()


def build_tree():
    root = Tnode("Productes")
    ele=Tnode("Electronics")
    vehical = Tnode("vehical")
    
    laptop = Tnode("Laptop")
    laptop.add_child(Tnode("Dell"))
    laptop.add_child(Tnode("HP"))
    laptop.add_child(Tnode("Apple"))
    
    phone = Tnode("Phone")
    phone.add_child(Tnode("Samsung"))
    phone.add_child(Tnode("Apple"))

    ele.add_child(laptop)
    ele.add_child(phone)

    car= Tnode("Car")
    bus = Tnode("Bus")
    truck = Tnode("Truck")

    vehical.add_child(car)
    vehical.add_child(bus)
    vehical.add_child(truck)
    
    root.add_child(ele)
    root.add_child(vehical)
   
    return root

def find_node(root, value):
    if root.value == value:
        return root
    for child in root.children:
        found = find_node(child, value)
        if found:
            return found
    return None

def insert_node(root, parent_value, new_value):
    parent_node = find_node(root, parent_value)
    if parent_node:
        new_node = Tnode(new_value)
        parent_node.add_child(new_node)
    else:
        print(f"Parent node with value '{parent_value}' not found.")

def delete_node(root,parent_value, value):
    parent_node = find_node(root, parent_value)
    if parent_node:
        for child in parent_node.children:
            if child.value == value:
                parent_node.children.remove(child)
                return
        print(f"Node with value '{value}' not found under parent '{parent_value}'.")
    else:
        print(f"Parent node with value '{parent_value}' not found.")



if __name__ == "__main__":
    root = build_tree()
    print("Tree structure:")
    root.print_tree()
    insert_node(root, "Electronics", "Tablet")
    insert_node(root,"Tablet","Samsung")
    insert_node(root,"Tablet","Apple")
    delete_node(root, "Electronics", "Phone")
    print("\nAfter inserting 'Tablet' under 'Electronics':")
    root.print_tree()
    

```
